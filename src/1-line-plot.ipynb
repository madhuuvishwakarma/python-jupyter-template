### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import matplotlib.pyplot as plt
import numpy as np

# Define x values ranging from 0 to 10
x = np.arange(0, 11, 1)
# Define y values as the square of x
y = x**2

# Create a line plot
plt.plot(x, y)

# Add title and labels for both axes
plt.title('Square of x')
plt.xlabel('x')
plt.ylabel('y = x^2')

# Add a grid
plt.grid(True)

# Display the plot
plt.show()
```
